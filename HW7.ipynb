**Solution:**

We have the linear regression likelihood:

$$
p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) \propto \exp \left(-\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top} \Sigma^{-1}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$

where $\Sigma=\sigma^2 I$, meaning the errors are i.i.d. normal with variance $\sigma^2$.
The prior on $\boldsymbol{\beta}$ follows a multivariate normal:

$$
p(\boldsymbol{\beta})=\mathcal{M} \mathcal{V} \mathcal{N}\left(\boldsymbol{\beta}_0, \Sigma_\beta\right)
$$

which implies:

$$
p(\boldsymbol{\beta}) \propto \exp \left(-\frac{1}{2}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)^{\top} \Sigma_\beta^{-1}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)\right)
$$


By Bayes' rule, the posterior is:

$$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) p(\boldsymbol{\beta})
$$


Substituting the likelihood and prior:

$$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto \exp \left(-\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top} \Sigma^{-1}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right) \times \exp \left(-\frac{1}{2}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)^{\top} \Sigma_\beta^{-1}\left(\boldsymbol{\beta}-\boldsymbol{\beta}_0\right)\right)
$$


Rearranging the exponent terms:

$$
-\frac{1}{2}\left[\boldsymbol{\beta}^{\top}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{X}+\Sigma_\beta^{-1}\right) \boldsymbol{\beta}-2 \boldsymbol{\beta}^{\top}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{y}+\Sigma_\beta^{-1} \boldsymbol{\beta}_0\right)\right]
$$

Recognizing the quadratic form of a multivariate normal density, we identify:

$$
\begin{gathered}
\Sigma_{\beta \mid y}=\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{X}+\Sigma_\beta^{-1}\right)^{-1} \\
\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma \sim \mathcal{M} \mathcal{V} \mathcal{N}\left(\mu_{\beta \mid y}, \Sigma_{\beta \mid y}\right)
\end{gathered}
$$

where:

$$
\mu_{\beta \mid y}=\Sigma_{\beta \mid y}\left(\mathbf{X}^{\top} \Sigma^{-1} \mathbf{y}+\Sigma_\beta^{-1} \boldsymbol{\beta}_0\right)
$$

This is the posterior distribution of $\beta$ given $\mathbf{y}, \mathbf{X}$, and $\Sigma$.

2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

**Solution:**

We assume:

1. The likelihood follows:

$$
\begin{gathered}
p\left(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2\right)=\mathcal{M} \mathcal{V} \mathcal{N}\left(\mathbf{X} \boldsymbol{\beta}, \sigma^2 I\right) \\
\propto\left(\sigma^2\right)^{-n / 2} \exp \left(-\frac{1}{2 \sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
\end{gathered}
$$

2. The prior on $\sigma^2$ is Inverse-Gamma:

$$
\begin{aligned}
& p\left(\sigma^2\right)=\operatorname{Inv-Gamma}\left(\alpha^*, \beta^*\right) \\
& \propto\left(\sigma^2\right)^{-\left(\alpha^*+1\right)} \exp \left(-\frac{\beta^*}{\sigma^2}\right)
\end{aligned}
$$


By Bayes' rule:

$$
p\left(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}\right) \propto p\left(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2\right) p\left(\sigma^2\right)
$$


Substituting the likelihood and prior:

$$
p\left(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}\right) \propto\left(\sigma^2\right)^{-n / 2} \exp \left(-\frac{1}{2 \sigma^2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right) \times\left(\sigma^2\right)^{-\left(\alpha^*+1\right)} \exp \left(-\frac{\beta^*}{\sigma^2}\right)
$$


Rearrange:

$$
p\left(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}\right) \propto\left(\sigma^2\right)^{-\left(\alpha^*+n / 2+1\right)} \exp \left(-\frac{\beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})}{\sigma^2}\right)
$$


This matches the Inverse-Gamma distribution form:

$$
\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \sim \operatorname{Inv-Gamma}\left(\alpha^*+\frac{n}{2}, \beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})\right)
$$


This is the posterior distribution of $\sigma^2$.

```python
import pymc3 as pm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import arviz as az
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_boston

# Load Boston Housing dataset
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target, name='MEDV')

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Bayesian Robust Regression Model
with pm.Model() as model:
    # Global precision parameter
    tau = pm.Gamma("tau", alpha=2, beta=2)
    
    # Local precision parameters for each observation (to handle outliers)
    tau_i = pm.Gamma("tau_i", alpha=2, beta=2, shape=len(y))
    
    # Regression coefficients
    beta = pm.Normal("beta", mu=0, sigma=1, shape=(X_scaled.shape[1],))
    intercept = pm.Normal("intercept", mu=0, sigma=1)
    
    # Likelihood function (accounting for individual tau_i values)
    sigma_sq = pm.Deterministic("sigma_sq", tau * tau_i)
    mu = intercept + pm.math.dot(X_scaled, beta)
    
    y_obs = pm.Normal("y_obs", mu=mu, sigma=pm.math.sqrt(1/sigma_sq), observed=y)
    
    # Sample from posterior
    trace = pm.sample(2000, tune=1000, return_inferencedata=True, cores=2)

# MCMC Diagnostics
az.plot_trace(trace)
plt.show()

# Posterior summary
print(az.summary(trace))

# Identifying outliers via posterior tau_i distributions
az.plot_posterior(trace, var_names=["tau_i"], kind='hist')
plt.show()
```
